````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as stats
````

> Statistical power is the probability of a hypothesis test of finding an effect if there is an effect to be found.

> Significance level (alpha): Boundary for specifying a statistically significant finding when interpreting the p-value.

> Type I Error. Reject the null hypothesis when there is in fact no significant effect (false positive). The p-value is optimistically small.

> Type II Error. Not reject the null hypothesis when there is a significant effect (false negative). The p-value is pessimistically large.

```r
Power = 1 - Type II Error
Pr(True Positive) = 1 - Pr(False Negative)
```

````python
plt.figure(figsize=(15,9))
plt.style.use('ggplot')



plt.suptitle('Statistical Power Illustration')

plt.title('Red Area - Type II Error \n Blue Area - Significance Level \n Green Area - Statistical Power')

mean=1000
std=250

x = np.linspace(mean-3*std, mean+3*std,1000)
z = np.linspace(mean+800-3*std, mean+800+3*std,1000)

iq = stats.norm(mean,std)
iq2 = stats.norm(mean+800,std)

plt.plot(x,iq.pdf(x),'b')
plt.plot(z,iq2.pdf(z),'g')

px=np.arange(1050,1500,10)
plt.fill_between(px,iq2.pdf(px),color='r')

px=np.arange(1500,1800,10)
plt.fill_between(px,iq.pdf(px),color='b')

px=np.arange(1500,2500,10)
plt.fill_between(px,iq2.pdf(px),linestyle = '--', color = 'g', alpha = 0.3)

plt.show()
````
